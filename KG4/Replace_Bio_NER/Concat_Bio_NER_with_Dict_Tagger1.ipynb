```python
# the function to get two and three combinations of entity words 

import re 

def get_two_three_comb(sentence):
    
        # remove punct 
    str1 = sentence.replace(',', '')
    str1 = str1.replace('.', '')
    
    # split sentence
    str1 = str1.split(' ')
    
    # for comb of two 
    comb_12 = []
    for i, j in enumerate(str1):

        comb12 = str1[i - 1] + " " + str1[i]
        comb_12.append(comb12)
        
    # for comb of three 
    comb_13 = []
    str1 = str1 + ['O']
    comb_12 = ['O'] + comb_12
    for i, j in zip(str1, comb_12): 
        comb13 = j + " "+ i
        comb_13.append(comb13)  
    
    # final list 
    comb123 = str1 + comb_12 + comb_13
    
    return comb123

# print(get_two_three_comb(words_corpus))

```

```python

def tag_for_a_sent(expanded_word_list, diseases_dict):
    
    sent_tagged = []
    for word in expanded_word_list:
        if word in diseases_dict: 
            sent_tagged.append(word)
            sent_tagged = list(sent_tagged)
    return sent_tagged 

test_1 = []
for sent in expanded_word_list: 
    temp = tag_for_a_sent(sent, diseases_dict)
    test_1.append(temp)
    
print(len(test_1))
print(test_1[1:20])
```
